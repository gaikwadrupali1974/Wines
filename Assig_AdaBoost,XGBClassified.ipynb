```python
# importing important libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
# Extracting data

from sklearn.datasets import load_wine
wine = load_wine()
print('Keys : ',wine.keys())
print('Keys : ', type(wine))


```

```python
# EDA

print(wine.DESCR)
```

```python
# Seperating the Input Data and Output Data

X = wine.data
y = wine.target

# performing trian test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2 , random_state = 5 )
```

```python
# Creating a Model

model = AdaBoostClassifier(n_estimators = 100, learning_rate = 0.001)

# Fitting the Model
model.fit(X_train, y_train)
```

```python
# Taking the prediction form the model on training and testing data

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# Getting the performance

print('Training Accuracy :', np.round(metrics.accuracy_score(y_train, y_train_pred), 3))
print('Training Precision:', np.round(metrics.precision_score(y_train, y_train_pred, average='weighted'), 3))
print('Training Recall :', np.round(metrics.recall_score(y_train, y_train_pred, average='weighted'), 3))
print('Training F1 score:', np.round(metrics.f1_score(y_train, y_train_pred, average='weighted'), 3))

print('\n Test Accuracy : ', np.round(metrics.accuracy_score(y_test, y_test_pred), 3))
print('Test Precision :',np.round(metrics.precision_score(y_test, y_test_pred, average='weighted'), 3))
print('Test Recall :', np.round(metrics.recall_score(y_test, y_test_pred,average='weighted'), 3))
print('Test F1 score:', np.round(metrics.f1_score(y_test, y_test_pred,average='weighted'), 3))
```

```python
'''
Comment -> The above model looks minimal as the performance on training and testing data is
(A) Below 90%
(B) Above 5% of each other
(C) This model is overfitting
'''
```
